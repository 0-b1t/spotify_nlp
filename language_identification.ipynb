```python
tracks['lang'] = None
```

```python


for i in range(247, tracks.shape[0]):
    song = tracks['lyrics'][i]
    if type(song)!=float:
        song= song.replace('\n', ' ').strip()
        blob = TextBlob(song)

        lang = blob.detect_language()
        tracks.loc[i,'lang'] = lang
    if i%50==0:
        print(i/tracks.shape[0])
    time.sleep(0.4)
```
