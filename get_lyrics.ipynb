```python
import time
```

```python
tracks["lyrics"] = None
```

```python
start = time.time()
for i in range(tracks.shape[0]):
    artist = tracks["artist"][i]
    title = tracks["title"][i]
    cleaned_title = re.sub(r"\([^)]*\)", "", title)
    cleaned_artist = re.sub(r"\([^)]*\)", "", artist)
    try:
        song = genius.search_song(cleaned_title, cleaned_artist)
        genius_title = song.title.replace("’", "'")
        genius_artist = song.artist.replace("’", "'")
        if ((genius_title in title) or (title in genius_title)) or (
            (genius_artist in artist) or (artist in genius_artist)
        ):
            print(i, "match")
            tracks.loc[i, "lyrics"] = song.lyrics
    except:
        time.sleep(5.1)
    if time.time() - start > 10 * 60:
        genius = lyricsgenius.Genius(API_KEY)
        start = time.time()
```
